```python
import pandas as pd 
import os
def convert_people_cell(cells):
    if cells =='n.a.':
        return 'gg'
    return cells

def convert_eps_cell(cells):
    if cells == "not available":
        return 'ktoo'
    return cells

df= pd.read_excel('stock_data.xlsx',sheet_name='Sheet1', 
           converters =  {'people':convert_people_cell, 'eps' : convert_eps_cell}
)
df
     
```

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stock_new.xlsx') as Writer:
    df_stocks.to_excel(Writer,sheet_name= 'stock')
    df_weather.to_excel(Writer,sheet_name= 'weather')
    
import os
    
df = pd.read_excel('stock_new_xlsx',sheet_name='stock')
df
```
